### Defining the vertices of obstacles, and points between vertices of the objects

```python
# Set vertices of the obstacles
CB1 = [(3,3),(3,4),(5,4),(5,3)]  # Rectangle
CB2 = [(7,4),(7,2),(8,2)]        # Triangle
CB = [(0,0),(10,0),(10,10),(0,10)] # Boundary

N = 50
M = 80
# Rectangle
pts1 = np.linspace(CB1[0], CB1[3], N)
pts2 = np.linspace(CB1[3], CB1[2], N)
pts3 = np.linspace(CB1[2], CB1[1], N)
pts4 = np.linspace(CB1[1], CB1[0], N)

# Triangle
pts5 = np.linspace(CB2[1], CB2[2], M)
pts6 = np.linspace(CB2[2], CB2[0], M)
pts7 = np.linspace(CB2[0], CB2[1], M)

# Border
pts8 = np.linspace(CB[0], CB[1], N)
pts9 = np.linspace(CB[1], CB[2], N)
pts10 = np.linspace(CB[2], CB[3], N)
pts11 = np.linspace(CB[3], CB[0], N)

obstacle_pts = np.concatenate((pts1,pts2,pts3,pts4,pts5,pts6,pts7,pts8,pts9,pts10,pts11))
```

```python
# ALL LINE SEGMENTS MAKING UP OBSTACLES
# Rectangle
l1 = (CB1[0], CB1[3])
l2 = (CB1[3], CB1[2])
l3 = (CB1[2], CB1[1])
l4 = (CB1[1], CB1[0])

# Triangle
l5 = (CB2[1], CB2[2])
l6 = (CB2[2], CB2[0])
l7 = (CB2[0], CB2[1])

# Border
l8 = (CB[0], CB[1])
l9 = (CB[1], CB[2])
l10 = (CB[2], CB[3])
l11 = (CB[3], CB[0])

lines = np.array([l1,l2,l3,l4,l5,l6,l7,l8,l9,l10,l11])
```

```python
mydist = potential_field(obstacle_pts=obstacle_pts,lines=lines,position=(2,1),qf=(9,4),step=.01,ro=.3)
mydist
```
